```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

driver = webdriver.Chrome('chromedriver',options=chrome_options)
products=[] #List to store name of the product
prices=[] #List to store price of the product
ratings=[] #List to store rating of the product
driver.get("https://www.amazon.in/s?k=hp+laptop&crid=P9LSTP6B1UYW&sprefix=hp+laptop%2Caps%2C232&ref=nb_sb_ss_ts-doa-p_2_9")
content = driver.page_source #get the page contents
soup = BeautifulSoup(content)
#CardInstance_otOB0z1ybgD6mGIR_6L4g > div._bXVsd_content_2rsXy

for a in soup.findAll('div', attrs={'class':'sg-col-inner'}):
  try:
      name=a.find('span', attrs={'class':'a-size-medium a-color-base a-text-normal'})
      print(name.text)
      price=a.find('span', attrs={'class':'a-price-whole'})
      print(price.text)
      rating=a.find('span', attrs={'class':'a-icon-alt'})
      print(rating.text)
      products.append(name.text)
      prices.append(price.text)
      ratings.append(rating.text)
  except AttributeError:
      continue
dflip = pd.DataFrame({'Product Name':products,'Price':prices,'Rating':ratings})
#dflip = pd.DataFrame({'Product Name':products,'Price':prices})
#dflip.to_csv('myWatch.csv', index=False, encoding='utf-8') 
```
