```python
from google_play_scraper import app
import pandas as pd
import numpy as np

#Indian Market

from google_play_scraper import Sort, reviews_all

reviews = reviews_all(
    'com.skin.rarebundle',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
print(reviews)
```

```python
df = pd.DataFrame(np.array(reviews), columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df.head()
```

# **App Reviews**
reviews function returns result with continuation token.

1. result : Crawling result of reviews. (list)
2. continuation_token : Data containing how many items were loaded, what arguments used in the current result. If you pass this value to the continuation_token parameter of the reviews function, the next items are crawled. For example, if 1000 reviews are retrieved and the returned token 'eXamplE' is passed to the reviews function, the list of reviews is retrieved from 1000 or later items.

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.fantome.penguinisle',
    lang='en', # defaults to 'en'
    country='in', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    count=100, # defaults to 100
    filter_score_with=5 # defaults to None(means all score)
)

# If you pass `continuation_token` as an argument to the reviews function at this point,
# it will crawl the items after 3 review items.
print(continuation_token)
result, _ = reviews(
    'com.fantome.penguinisle',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
```

```python
print(result)
```
