```python
import calendar
yy = 2022
mm = 4

#display the specific month of a year
print(calendar.month(yy,mm))
```

```python
print(calendar.calendar(2022, 2,1,6))
```

```python
from datetime import date, timedelta

def allsundays(year):
  d = date(year,1,1)
  d +=timedelta(days=6-d.weekday())  #will get you the first sunday
  while d.year ==year:
    yield d
    d += timedelta(days=7)

for d in allsundays(2022):
  print(d)


```

# **Rock, Paper, Scissors**

```python
import random
user_choice = input("Enter a choice (rock, paper, scissors): ")
possible_action =["rock", "paper", "scissors"]
computer_action = random.choice(possible_action)
print(f"\nYou chose {user_choice}, computer chose {computer_action}.\n")

if user_choice == computer_action:
  print(f"Both players selected {user_choice}. It's a tie!")
elif user_choice=="rock":
  if computer_action=="scissors":
    print("You smashed the computer !!!  You win !!")
  else:
    print("You are covered by the computer !!!  You lose !!")
elif user_choice=="paper":
  if computer_action=="rock":
    print("You covered the computer !!!  You win !!")
  else:
    print("You are cut by the computer !!!  You lose !!")  
elif user_choice=="scissors":
  if computer_action=="paper":
    print("You cut the computer !!!  You win !!")
  else:
    print("You are smashed by the computer !!!  You lose !!")
```

```python
from imageai.Detection import ObjectDetection
detector = ObjectDetection()   #create an object
model_path = "yolo-tiny.h5"
input_path = "testTraffic.jpeg"
output_path = "newimage.jpg"
```

```python
detector.setModelTypeAsTinyYOLOv3()
detector.setModelPath(model_path)
detector.loadModel()
detection = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path)

for eachItem in detection:
    print(eachItem["name"] , " : ", eachItem["percentage_probability"])
```
