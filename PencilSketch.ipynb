# **Pencil Sketch using Python**

```python
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread("anuj.jpeg")
gray_scale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#invert the image
inverted_image = 255-gray_scale
blurred = cv2.GaussianBlur(inverted_image, (21,21), 0)
inverted_blur = 255- blurred
pencil_sketch= cv2.divide(gray_scale, inverted_blur, scale=256.0)
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
